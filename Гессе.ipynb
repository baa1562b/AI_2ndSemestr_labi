```
from sympy import *
x1 = symbols('x1')
x2 = symbols('x2')
#y = 10*x*log(x)-(x**2)/2
# f = x1 * x2 - 1/(2*(x1+x2))
#f = exp(2*x1) * (x1 + x2**2 - 2*x2)
#f = x1**2 - 2*x1*x2-x2**2
f = 2 * x1**2 +x1*x2 -x2**2-8*x1+8*x2
```

```
grad = [diff(f, i) for i in [x1, x2]]
hess = hessian(f, (x1,x2))
# solvers.solve(grad)
[XC1, XC2] = [solvers.solve(grad)[x1],solvers.solve(grad)[x2]]
print(XC1, XC2)
hess_resh = hess.subs(x1, XC1).subs(x2,XC2)
print(hess_resh)
```

```
if hess_resh.minor(1,1)>0 and hess_resh.det()>0:
  print("Локальный минимум %f, %f, %f" % (XC1, XC2, f.subs(x1,XC1).subs(x2,XC2)))
if hess_resh.minor(1,1)<0 and hess_resh.det()>0:
  print("Локальный максимум %f, %f, %f" % (XC1, XC2, f.subs(x1,XC1).subs(x2,XC2)))
if hess_resh.det()<0:
  print("Нет решения")
```
